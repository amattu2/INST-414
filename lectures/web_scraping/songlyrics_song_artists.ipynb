```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "http://www.songlyrics.com/a"

page = requests.get(url)
page.status_code
```

```python
soup = bs(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
pages = soup.find_all(class_= "li_pagination")
```

```python
pages[264].a.get("href")
```

```python
urls_a = ["http://www.songlyrics.com" + item.a.get("href") for item in pages[1:len(pages)]]
```

```python
page = requests.get(urls_a[102])
page.status_code
soup = bs(page.content, "html.parser")
```
