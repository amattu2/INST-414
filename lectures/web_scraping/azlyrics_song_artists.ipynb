```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.azlyrics.com/a.html"

page = requests.get(url)
page.status_code
```

```python
soup = bs(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
artists = soup.find_all(class_= "col-sm-6 text-center artist-col")
```

```python
artists_left = artists[0].find_all("a")
artists_right = artists[1].find_all("a")
```

```python
links = []

for column in artists:
    a_tags = column.find_all("a")
    for item in a_tags:
        links.append(item.get("href"))
```

```python
links
```

```python
fqdn_urls = ["https://azlyrics.com/" + addition for addition in links]
```
