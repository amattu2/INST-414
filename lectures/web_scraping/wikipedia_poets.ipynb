```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_poets")
```

```python
page.status_code
```

```python
soup = bs(page.content, "lxml")
```

```python
soup.title
```

```python
list_items = soup.find_all("li")
```

```python
print(list_items)
```

```python
list_items[356].text
```

```python
list_items[356].a
```

```python
list_items[356].a.get("href")
```

```python
links = []

for item in list_items:
    if (item.a is None):
        continue
    url = item.a.get("href")
    links.append(url)
```

```python
print(links)
```

```python
texts = []

for item in list_items:
    if (item.text is None):
        continue
    text = item.text
    texts.append(text)
```

```python
texts
```

```python
data = list(zip(*(texts, links)))
```

```python
data[256]
```

```python
import csv

with open("output.csv", "w", encoding = "utf8", newline = "") as file:
    wr = csv.writer(file)
    wr.writerow(("Text", "Link"))
    wr.writerow(data)
```
